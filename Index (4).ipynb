```python
start = 'W'
endereco_mp = {}
endereco_cache = {}
for place in range(0, 16):
    (endereco_mp['{0:04b}'.format(place)]) = '00000000'
for place1 in range(0, 2):
    (endereco_cache['Linha {}'.format(place1)]) = {}
    for place2 in range(0,4):
         (endereco_cache['Linha {}'.format(place1)]['Bloco {0:03b}'.format(place2)]) = None
while start == 'W' or start == 'L' or start == 'R':
    start = input('Digite W para escrever, R para ler, L para listar toda a memoria e qualquer tecla para parar. ')
    if start == 'W':
        local = input('Digite o endereco de 4 bits: ')
        palavra = input('Digite o dado de 8 bits: ')
        endereco_mp[local] = palavra
    elif start == 'R':
        local = input('Digite o endereco: ')
        for _ in range(0,2):
            tag = endereco_cache['Linha {}'.format(_)]['Bloco {0:03b}'.format(_)]
        if tag == endereco_mp.get(local):
            print(endereco_cache[local])
        else:
            if local == '0000' or local == '0100' or local == '1000':
                endereco_cache['Linha 0']['Bloco {}'.format(range(0,4))] = endereco_mp.get(local)
            elif local == '0001' or local == '0101' or local =='1001':
                endereco_cache['Linha 1']['Bloco {}'.format(range(0,4))] = endereco_mp.get(local)
            elif local == '0010' or local == '0110' or local == '1010':
                endereco_cache['Linha 0']['Bloco {}'.format(range(0,4))] = endereco_mp.get(local)
            elif local == '0011' or local == '0111' or local == '1111':
                endereco_cache['Linha 1']['Bloco {}'.format(range(0,4))] = endereco_mp.get(local)
            print(endereco_mp.get(local))
    elif start == 'L':
        print('Todos os estados: ')
        for element in endereco_mp:
            print(endereco_mp[element])
```
